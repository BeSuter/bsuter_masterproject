```python
import numpy as np
import healpy as hp

NSIDE = 32
NPIX = hp.nside2npix(NSIDE)
```

```python
#Try to make sphere with random values...
all_ipix = list(range(NPIX))
m = np.arange(NPIX)
```

```python
mean_value = 12
for pixel in all_ipix:
    theta, phi = np.degrees(hp.pix2ang(nside=32, ipix=[pixel]))
    vec = hp.ang2vec(np.radians(theta), np.radians(phi))[0]
    ipix_disc = hp.query_disc(nside=32, vec=vec, radius=np.radians(10))
    m[ipix_disc] = np.random.normal(loc=mean_value)
hp.mollview(m, title="Mollview image RING with mean=%d" % mean_value)
hp.graticule()
```
